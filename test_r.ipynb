## **Data Import using R**
---
Code to import publicly available dataset using R packages  

```r
#install.packages(c("xts","lubridate","crypto2"))
```

```r
library(data.table)
library(lubridate)
library(crypto2)
library(xts)
```

```r
full_list = crypto_list();
```

```r
coins = crypto_history(coin_list = full_list[2,],
                       start_date = "20150101",
                       end_date = "20221231",
                       interval = "1d")
LTC = data.table(date = as.Date(substr(coins$timestamp,1,10), format = '%Y-%m-%d'),
                 open = coins$open,
                 high = coins$high,
                 low = coins$low,
                 close = coins$close)
```
